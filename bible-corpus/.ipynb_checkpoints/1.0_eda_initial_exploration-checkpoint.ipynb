```python
sns.set(font_scale=3.1)

plt.figure(figsize=(60, 45))

sns.barplot(x='b', y='p', data=key_abbrev, ci=False)
plt.xlabel('Book ID', fontsize=70)
plt.ylabel('Abbreviation Count', fontsize=70)
plt.title(label='Bible Book Abbreviation Count by ID Number', fontsize=80);
```

#### This graph visualizes the book IDs for both the old and new testament in chronological order from top to bottom.

```python
sns.set(font_scale=3.3)

plt.figure(figsize=(29, 95))
sns.barplot(x='b', y='a', data=key_abbrev);           
```

```python
sns.set(font_scale=0)
plt.figure(figsize=(300, 200))
sns.barplot(x='a', y='b', data=key_abbrev)
plt.xlabel('Book Names Ordered From The Old to The New Testament', fontsize=400)
plt.ylabel('Count of Book ID', fontsize=400)
plt.title(label='Bible Book Abbreviation Count by ID Number', fontsize=480);        
```
